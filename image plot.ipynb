```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#prepare data
img=Image.open("image.png")

#convert to .npy
arr=np.array(img)

#Add figure
fig=plt.figure(figsize=(10,5))

#Add axes
ax1=plt.subplot()

#plot image
ax1.imshow(arr)
```

# image plot to histogram

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#prepare data
img=Image.open("image.png")

#convert to .npy
arr=np.array(img)

#Add figure
fig=plt.figure(figsize=(10,5))

#Add axes
ax1=plt.subplot()

#plot the histogram of this image
ax1.hist(arr.ravel(),bins=20,range=(0,260),fc='orange',ec='blue')  
#arr.ravel()...returns contigous flattened array (1D array withall the input array elements and with the same type as it display the histogram)

plt.show()
```
