```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

# define the dataset 
x=range(1,15)
y=[1,4,6,8,4,5,3,2,4,1,5,6,8,7]

#creating the area plot
plt.stackplot(x,y,colors='Black',alpha=0.5)

#merge area chart with the line chart
plt.plot(x,y,color='g')
plt.grid(True)
plt.show()
```

```python
#define the dataset
x=np.arange(0,10,1)
y=2*x+5

#creating the area plot
plt.stackplot(x,y,colors='Blue',alpha=0.5)

#merge area chart with the line chart
plt.plot(x,y,color='r')
plt.grid(True)
plt.show()
```
