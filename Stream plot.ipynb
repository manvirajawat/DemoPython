```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# prepare data
x=np.arange(0,10)
y=np.arange(0,10)
X,Y=np.meshgrid(x,y)
u=np.ones((10,10))  # x-component to the right
v=np.zeros((10,10)) # y-component zero

# add figure
fig=plt.figure(figsize=(10,5))

# add axes
ax1=plt.subplot()

# plot
ax1.streamplot(X,Y,u,v,density=0.5)

# customization title
plt.title('Stream Plot 1')

# save the plot
plt.savefig('StreamPlot1.png')

#show plot
plt.show()
```
