```python
import matplotlib.pyplot as plt
%matplotlib inline

label=['Frog','Mogs','Dogs','Logs']
sizes=[15,30,45,10]

fig=plt.figure(figsize=(10,5))

ax1=plt.subplot()

ax1.pie(sizes,labels=label,autopct='%1.1f%%',shadow=True,startangle=90)

plt.title('Pie Chart')

plt.savefig('Piechart.png')

plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

label=['Dog','Cat','Wolf','Lion']
sizes=[50,45,60,80]

fig=plt.figure(figsize=(10,5))

ax1=plt.subplot()

ax1.pie(sizes,labels=label,autopct='%1.1f%%',shadow=True,startangle=90,explode=(0.1,0.1,0,0)

plt.title('Pie Chart')

plt.savefig('Piechart1.png')

plt.show()
```
